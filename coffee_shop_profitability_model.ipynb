# Coffee Shop Profitability Model

This project builds a simple financial model for a small coffee shop to estimate revenue, expenses, profit, and break-even points using Python.

## Objective

This project analyzes the financial performance of a hypothetical coffee shop using a simple profitability model. The goal is to estimate revenue, expenses, profit margins, and the break-even point.

## Business Assumptions

The financial model uses the following assumptions:

- Average coffee price: $5
- Customers per day: 120
- Days open per month: 30

## Revenue Model

```python
price = 5
customers_per_day = 120
days_per_month = 30

monthly_revenue = price * customers_per_day * days_per_month

print("Monthly Revenue:", monthly_revenue) 
```

## Expense Model

```python
rent = 3000
utilities = 600
equipment = 400
insurance = 200

coffee_beans = 2500
milk_supplies = 1200
cups = 800
wages = 6000

total_expenses = rent + utilities + equipment + insurance + coffee_beans + milk_supplies + cups + wages

total_expenses
```

## Profit Calculation

```python
profit = monthly_revenue - total_expenses
profit 
```

## Break-even Analysis

```python
break_even_customers_month = total_expenses / price
break_even_customers_day = break_even_customers_month / days_per_month

break_even_customers_day
```

## Visualization

```python
import matplotlib.pyplot as plt

labels = ['Revenue', 'Expenses', 'Profit']
values = [monthly_revenue, total_expenses, profit]

plt.bar(labels, values)
plt.title("Coffee Shop Financial Performance")
plt.ylabel("USD")
plt.show() 
```

## Conclusion

This financial model suggests that the coffee shop would generate approximately $18,000 in monthly revenue under the current assumptions.

After accounting for operating costs such as rent, wages, supplies, and utilities, the business produces an estimated monthly profit of about $3,300.

The break-even analysis indicates the shop must serve approximately 98 customers per day to cover its expenses. Since the model assumes 120 daily customers, the business operates with a margin of safety.

Overall, the model demonstrates that the coffee shop could be financially viable under the given assumptions.
