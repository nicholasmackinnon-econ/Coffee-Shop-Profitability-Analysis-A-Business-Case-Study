# Coffee Shop Profitability Model

This project builds a simple financial model for a small coffee shop to estimate revenue, expenses, profit, and break-even points using Python.

## Objective

This project analyzes the financial performance of a hypothetical coffee shop using a simple profitability model. The goal is to estimate revenue, expenses, profit margins, and the break-even point.

## Business Assumptions

The financial model uses the following assumptions:

- Average coffee price: $5
- Customers per day: 120
- Days open per month: 30

# Financial Model

This section defines the assumptions and calculations used to estimate revenue, expenses, profit, and break-even levels for the coffee shop.

## Revenue Model

```python
price = 5
customers_per_day = 120
days_per_month = 30

monthly_revenue = price * customers_per_day * days_per_month

print("Monthly Revenue:", monthly_revenue) 
```

## Expense Model

```python
rent = 3000
utilities = 600
equipment = 400
insurance = 200

coffee_beans = 2500
milk_supplies = 1200
cups = 800
wages = 6000

total_expenses = rent + utilities + equipment + insurance + coffee_beans + milk_supplies + cups + wages

total_expenses
```

## Profit Calculation

```python
profit = monthly_revenue - total_expenses
profit 
```

## Break-even Analysis

```python
break_even_customers_month = total_expenses / price
break_even_customers_day = break_even_customers_month / days_per_month

break_even_customers_day
```

# Scenario Analysis

This section tests how changes in customer volume and pricing affect profitability.

## Scenario Analysis: Customer Volume Impact

```python
customers_list = [80, 100, 120, 140]
price = 5
days = 30
monthly_expenses = total_expenses  # uses your existing variable

print("Customer Scenario Analysis:\n")

for customers in customers_list:
    revenue = customers * price * days
    profit = revenue - monthly_expenses
    print(f"Customers/day: {customers} → Monthly Profit: ${profit}")
```

## Scenario Analysis: Price Impact

```python
prices = [4, 5, 6]

print("\nPrice Scenario Analysis:\n")

for p in prices:
    revenue = 120 * p * 30
    profit = revenue - monthly_expenses
    print(f"Price: ${p} → Monthly Profit: ${profit}")
```

# Data Visualization

This section presents visual representations of the coffee shop’s financial performance and how profitability changes with customer demand.

## Visualization: Financial Performance

```python
import matplotlib.pyplot as plt

labels = ['Revenue', 'Expenses', 'Profit']
values = [monthly_revenue, total_expenses, profit]

plt.bar(labels, values)
plt.title("Coffee Shop Financial Performance")
plt.ylabel("USD")
plt.show() 
```

### Interpretation

The chart shows that the coffee shop generates strong revenue relative to its operating costs, resulting in a positive monthly profit of approximately $3,300. 

However, the relatively small gap between revenue and expenses indicates that profit margins are not very large, reinforcing the earlier insight that the business is sensitive to changes in customer demand and costs.

## Visualization: Profit vs Customers

```python
import matplotlib.pyplot as plt

customers_list = [80, 100, 120, 140]
profits = []

for customers in customers_list:
    revenue = customers * 5 * 30
    profit = revenue - total_expenses
    profits.append(profit)

plt.plot(customers_list, profits, marker='o')
plt.title("Profit vs Daily Customers")
plt.xlabel("Customers per Day")
plt.ylabel("Monthly Profit ($)")
plt.grid()

plt.show()
```

### Interpretation

The chart shows a clear relationship between customer volume and profitability. As the number of daily customers increases, monthly profit rises significantly, while lower customer levels quickly lead to losses.

The break-even point occurs at approximately 98 customers per day, where profit transitions from negative to positive. This reinforces that the business is highly sensitive to customer demand, and maintaining sufficient daily traffic is critical for profitability.
